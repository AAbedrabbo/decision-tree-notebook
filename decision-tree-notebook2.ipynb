```python
import pandas as pd
from sklearn import tree
```

```python
data = pd.read_csv("HR_clean.csv")
data.head()
```

```python
%matplotlib notebook
plot1 = data.plot.scatter(x = 'last_evaluation', y = 'satisfaction_level', c = 'resigned', colormap = 'cool')
```

```python
Y = data["resigned"]
X = data.iloc[:,4:-1]
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
clf
```

```python
!pip3 install graphviz
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render("data") 
```
